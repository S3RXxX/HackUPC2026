```python
import pandas as pd

def merge_malpractices(csv1_path, csv2_path, csv3_path, output_path="malpractices.csv"):
    # Load datasets
    df1 = pd.read_csv(csv1_path)
    df2 = pd.read_csv(csv2_path)
    df3 = pd.read_csv(csv3_path)

    # Extract required columns (with safety checks)
    col1 = df1["description"] if "description" in df1.columns else pd.Series([])
    col2 = df2["description"] if "description" in df2.columns else pd.Series([])
    col3 = df3["Message"] if "Message" in df3.columns else pd.Series([])

    # Normalize into single column DataFrames
    out1 = pd.DataFrame({"Malpractice": col1})
    out2 = pd.DataFrame({"Malpractice": col2})
    out3 = pd.DataFrame({"Malpractice": col3})

    # Combine all
    result = pd.concat([out1, out2, out3], ignore_index=True)

    # Drop empty / NaN rows
    result = result.dropna().reset_index(drop=True)

    # Save
    result.to_csv(output_path, index=False)

    print(f"Saved merged file to: {output_path}")
    return result
```
